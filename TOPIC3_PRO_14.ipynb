```python
def check_subset_sum(arr, target):
    n = len(arr)

    # Iterate through all possible subsets using bit masks
    # (1 << n) generates 2^n, which is the total number of subsets
    for mask in range(1, 1 << n): # Start from 1 to avoid empty set
        current_sum = 0
        # Check each element to see if it's in the current subset
        for i in range(n):
            # If the i-th bit is set in the mask, include arr[i] in the sum
            if (mask >> i) & 1:
                current_sum += arr[i]

        # If the current subset's sum matches the target, we found it
        if current_sum == target:
            return True

    # If no subset sums to the target after checking all possibilities
    return False

# Example usage:
arr = [1, 3, 9, 2, 7, 12]
target = 15

if check_subset_sum(arr, target):
    print("true")
else:
    print("false")
```
