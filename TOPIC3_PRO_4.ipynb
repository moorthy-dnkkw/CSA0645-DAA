```python
%%bash

cat > merge_sort.c << EOF
#include <stdio.h>

int count = 0;

void merge(int arr[], int l, int m, int r) {
    int n1 = m-l+1;
    int n2 = r-m;

    int L[n1], R[n2];

    for(int i=0;i<n1;i++) L[i]=arr[l+i];
    for(int j=0;j<n2;j++) R[j]=arr[m+1+j];

    int i=0,j=0,k=l;

    while(i<n1 && j<n2){
        count++;
        if(L[i]<=R[j])
            arr[k++]=L[i++];
        else
            arr[k++]=R[j++];
    }

    while(i<n1) arr[k++]=L[i++];
    while(j<n2) arr[k++]=R[j++];
}

void mergeSort(int arr[], int l, int r){
    if(l<r){
        int m=(l+r)/2;
        mergeSort(arr,l,m);
        mergeSort(arr,m+1,r);
        merge(arr,l,m,r);
    }
}

int main(){
    int arr[]={12,4,78,23,45,67,89,1};
    int n=8;

    mergeSort(arr,0,n-1);

    for(int i=0;i<n;i++)
        printf("%d ",arr[i]);

    printf("\nComparisons = %d",count);

    return 0;
}
EOF

gcc merge_sort.c -o merge_sort
./merge_sort
```
