```python
def find_kth_positive(arr, k):
    missing = []
    num = 1
    i = 0

    while len(missing) < k:
        if i < len(arr) and arr[i] == num:
            i += 1
        else:
            missing.append(num)
        num += 1

    return missing[-1]

print(find_kth_positive([2,3,4,7,11], 5))
print(find_kth_positive([1,2,3,4], 2))
```
