```python
%%bash
cat <<EOF > closest_sum.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int arr[] = {45,34,4,12,5,2};
    int n = 6;
    int target = 42;

    int closest = arr[0];

    for(int mask=1; mask<(1<<n); mask++) {
        int sum = 0;

        for(int i=0;i<n;i++) {
            if(mask & (1<<i))
                sum += arr[i];
        }

        if(abs(target - sum) < abs(target - closest))
            closest = sum;
    }

    printf("Closest Sum = %d", closest);

    return 0;
}
EOF
gcc -o closest_sum closest_sum.c
./closest_sum
```
