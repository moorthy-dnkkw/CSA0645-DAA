```python
def findPaths(m,n,N,i,j):
    MOD = 10**9+7
    dp = {(i,j):1}

    count = 0

    for _ in range(N):
        temp = {}

        for (x,y),ways in dp.items():
            for dx,dy in [(1,0),(-1,0),(0,1),(0,-1)]:
                nx,ny = x+dx,y+dy

                if nx<0 or nx>=m or ny<0 or ny>=n:
                    count=(count+ways)%MOD
                else:
                    temp[(nx,ny)] = temp.get((nx,ny),0)+ways

        dp=temp

    return count

print(findPaths(2,2,2,0,0))

```
