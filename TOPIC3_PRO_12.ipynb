```python
%%bash

cat > solution.c << EOF
#include <stdio.h>
#include <stdlib.h>

int compare(const void *a, const void *b) {
    return (*(int*)a - *(int*)b);
}

int main() {
    int arr[] = {23,17,31,44,55,21,20,18,19,27};
    int n = 10, k = 5;

    qsort(arr, n, sizeof(int), compare);

    printf("%d\n", arr[k-1]);

    return 0;
}
EOF

gcc solution.c -o solution
./solution
```
