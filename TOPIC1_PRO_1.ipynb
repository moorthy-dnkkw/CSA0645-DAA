```python
def rob(nums):

    def helper(arr):
        prev = curr = 0

        for num in arr:
            prev,curr = curr,max(curr,prev+num)

        return curr

    if len(nums)==1:
        return nums[0]

    return max(helper(nums[:-1]),helper(nums[1:]))

print(rob([2,3,2]))
```
