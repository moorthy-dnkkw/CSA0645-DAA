```python
def findIntersectionValues(nums1, nums2):
    answer1 = sum(1 for x in nums1 if x in nums2)
    answer2 = sum(1 for x in nums2 if x in nums1)
    return [answer1, answer2]

print(findIntersectionValues([2,3,2],[1,2]))
```
