```python
%%writefile main.c
#include <stdio.h>

int main() {
    int A[] = {1,2};
    int B[] = {-2,-1};
    int C[] = {-1,2};
    int D[] = {0,2};

    int count = 0;

    for(int i=0;i<2;i++) {
        for(int j=0;j<2;j++) {
            for(int k=0;k<2;k++) {
                for(int l=0;l<2;l++) {
                    if(A[i]+B[j]+C[k]+D[l] == 0)
                        count++;
                }
            }
        }
    }

    printf("Output = %d", count);

    return 0;
}

!gcc main.c -o main
!./main
```
