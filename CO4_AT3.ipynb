```python
import heapq

# Bellman-Ford Algorithm
def bellman_ford(vertices, edges, source):
    dist = [float('inf')] * vertices
    dist[source] = 0

    for _ in range(vertices - 1):
        for u, v, w in edges:
            if dist[u] != float('inf') and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w

    return dist

# Dijkstra Algorithm
def dijkstra(vertices, graph, source):
    dist = [float('inf')] * vertices
    dist[source] = 0

    pq = [(0, source)]

    while pq:
        d, u = heapq.heappop(pq)

        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    return dist

vertices = 5

edges = [
    (0,1,-1),
    (0,2,4),
    (1,2,3),
    (1,3,2),
    (1,4,2),
    (3,1,1),
    (4,3,-3)
]

graph = [[] for _ in range(vertices)]
for u, v, w in edges:
    graph[u].append((v, w))

print("Bellman-Ford:", bellman_ford(vertices, edges, 0))
print("Dijkstra:", dijkstra(vertices, graph, 0))
```

```python
import heapq

# Optimal Binary Search Tree Cost (DP)
def obst(freq):
    n = len(freq)
    cost = [[0] * n for _ in range(n)]

    for i in range(n):
        cost[i][i] = freq[i]

    for L in range(2, n + 1):
        for i in range(n - L + 1):
            j = i + L - 1
            cost[i][j] = float('inf')

            total = sum(freq[i:j+1])

            for r in range(i, j + 1):
                c = total

                if r > i:
                    c += cost[i][r - 1]

                if r < j:
                    c += cost[r + 1][j]

                cost[i][j] = min(cost[i][j], c)

    return cost[0][n - 1]

# Huffman Coding Cost (Greedy)
def huffman_cost(freq):
    heap = freq[:]
    heapq.heapify(heap)

    total_cost = 0

    while len(heap) > 1:
        x = heapq.heappop(heap)
        y = heapq.heappop(heap)

        merged = x + y
        total_cost += merged

        heapq.heappush(heap, merged)

    return total_cost

freq = [10, 20, 30, 40]

print("OBST Cost =", obst(freq))
print("Huffman Cost =", huffman_cost(freq))
```
