```python
def bin_packing(items, capacity):

    bins = []

    def backtrack(index):

        if index == len(items):
            print("Bin Assignments:")
            for i, b in enumerate(bins):
                print("Bin", i + 1, ":", b)
            return True

        item = items[index]

        # Try placing item in existing bins
        for i in range(len(bins)):

            if sum(bins[i]) + item <= capacity:

                bins[i].append(item)

                if backtrack(index + 1):
                    return True

                bins[i].pop()      # Backtrack

        # Create a new bin
        bins.append([item])

        if backtrack(index + 1):
            return True

        bins.pop()                # Backtrack

        return False

    backtrack(0)

# Example
items = [4, 8, 1, 4, 2, 1]
capacity = 10

bin_packing(items, capacity)
```

```python
def exact_cover(universe, subsets):

    solution = []

    def backtrack(covered):

        # All elements covered exactly once
        if covered == universe:
            return True

        # Select an uncovered element
        remaining = universe - covered
        element = next(iter(remaining))

        for subset in subsets:

            if element in subset:

                # Pruning: avoid overlap
                if covered.intersection(subset):
                    continue

                solution.append(subset)

                if backtrack(covered.union(subset)):
                    return True

                solution.pop()     # Backtrack

        return False

    if backtrack(set()):
        return solution
    else:
        return None

# Example
U = {1, 2, 3, 4}

S = [
    {1, 2},
    {3},
    {4},
    {2, 3},
    {1, 4}
]

result = exact_cover(U, S)

print("Exact Cover =", result)
```
