```python
def min_max(arr, low, high):
    # Base case 1: Only one element in the array
    if low == high:
        return (arr[low], arr[low])

    # Base case 2: Two elements in the array
    if high == low + 1:
        if arr[low] < arr[high]:
            return (arr[low], arr[high])
        else:
            return (arr[high], arr[low])

    # Recursive step: Divide the array into two halves
    mid = (low + high) // 2
    left_min, left_max = min_max(arr, low, mid)
    right_min, right_max = min_max(arr, mid + 1, high)

    # Combine the results from the two halves
    overall_min = min(left_min, right_min)
    overall_max = max(left_max, right_max)

    return (overall_min, overall_max)

# Main part of the program
if __name__ == "__main__":
    arr = [5, 7, 3, 4, 9, 12, 6, 2]
    n = len(arr)

    min_val, max_val = min_max(arr, 0, n - 1)

    print(f"Min = {min_val}")
    print(f"Max = {max_val}")
```
