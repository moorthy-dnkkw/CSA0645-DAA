```python
%%bash

# Create the C source file
cat << 'EOF' > my_c_program.c
#include <stdio.h>
#include <stdlib.h>

int compare(const void *a, const void *b) {
    return (*(int*)a - *(int*)b);
}

int main() {
    int arr[] = {12,3,5,7,19};
    int n = 5, k = 2;

    qsort(arr, n, sizeof(int), compare);

    printf("%d\n", arr[k]);

    return 0;
}
EOF

# Compile and run the C program
gcc -o my_c_program my_c_program.c
./my_c_program
```
