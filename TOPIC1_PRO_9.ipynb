```python
def binarySearch(arr, key):
    arr.sort()

    low = 0
    high = len(arr)-1

    while low <= high:
        mid = (low + high)//2

        if arr[mid] == key:
            return mid + 1

        elif arr[mid] < key:
            low = mid + 1

        else:
            high = mid - 1

    return -1

arr = [3,4,6,-9,10,8,9,30]
key = 10

pos = binarySearch(arr,key)

if pos != -1:
    print("Element found at position",pos)
else:
    print("Element not found")

```
