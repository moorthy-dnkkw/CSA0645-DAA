```python
arr = [5, 10, 15, 20, 25, 30, 35, 40, 45]
n = len(arr)
key = 20
low = 0
high = n - 1
comparisons = 0
found = False

while low <= high:
    comparisons += 1
    mid = (low + high) // 2  # Use integer division

    if arr[mid] == key:
        print(f"Position = {mid + 1}")
        print(f"Comparisons = {comparisons}")
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1

if not found:
    print("Not Found")
```
