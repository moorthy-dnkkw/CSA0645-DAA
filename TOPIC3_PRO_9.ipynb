```python
%%writefile main.c
#include <stdio.h>
#include <math.h>

int main() {
    int points[4][2] = {{1,3},{-2,2},{5,8},{0,1}};
    int n = 4, k = 2;

    float dist[4];

    for(int i=0;i<n;i++)
        dist[i] = sqrt(points[i][0]*points[i][0] + points[i][1]*points[i][1]);

    for(int i=0;i<n-1;i++) {
        for(int j=i+1;j<n;j++) {
            if(dist[i] > dist[j]) {
                float temp = dist[i];
                dist[i] = dist[j];
                dist[j] = temp;

                int x = points[i][0];
                int y = points[i][1];

                points[i][0] = points[j][0];
                points[i][1] = points[j][1];

                points[j][0] = x;
                points[j][1] = y;
            }
        }
    }

    for(int i=0;i<k;i++)
        printf("[%d,%d] ", points[i][0], points[i][1]);

    return 0;
}
```
