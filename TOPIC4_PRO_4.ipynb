```python
%%bash
# Write the C code to a file
cat > tsp.c <<EOF
#include <stdio.h>
#include <limits.h>

int graph[4][4] = {
    {0,10,15,20},
    {10,0,35,25},
    {15,35,0,30},
    {20,25,30,0}
};

int visited[4];
int ans = INT_MAX;

void tsp(int city, int count, int cost) {
    if(count==4){
        cost += graph[city][0];
        if(cost<ans) ans=cost;
        return;
    }

    for(int i=0;i<4;i++){
        if(!visited[i]){
            visited[i]=1;
            tsp(i,count+1,cost+graph[city][i]);
            visited[i]=0;
        }
    }
}

int main() {
    visited[0]=1;
    tsp(0,1,0);
    printf("%d",ans);
}
EOF

# Compile the C code
gcc tsp.c -o tsp

# Run the compiled program
./tsp
```
