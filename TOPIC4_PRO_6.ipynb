```python
%%bash
cat << EOF > temp.c
#include <stdio.h>
#include <string.h>

int main() {
    char s[]="babad";
    int n=strlen(s);
    int start=0,maxLen=1;

    for(int i=0;i<n;i++){
        int l=i,r=i;
        while(l>=0&&r<n&&s[l]==s[r]){
            if(r-l+1>maxLen){
                start=l;
                maxLen=r-l+1;
            }
            l--; r++;
        }

        l=i; r=i+1;
        while(l>=0&&r<n&&s[l]==s[r]){
            if(r-l+1>maxLen){
                start=l;
                maxLen=r-l+1;
            }
            l--; r++;
        }
    }

    for(int i=start;i<start+maxLen;i++)
        printf("%c",s[i]);

    return 0;
}
EOF

gcc temp.c -o temp
./temp
```
