```python
%%bash
# Write the C code to a file
cat <<EOF > solution.c
#include <stdio.h>
#include <string.h>

int main() {
    char s[]="abcabcbb";
    int n=strlen(s), max=0;

    for(int i=0;i<n;i++){
        int visited[256]={0};
        for(int j=i;j<n;j++){
            if(visited[s[j]]) break;
            visited[s[j]]=1;
            if(j-i+1>max) max=j-i+1;
        }
    }

    printf("%d",max);
    return 0; // Added return 0 for good practice
}
EOF

# Compile the C code
gcc solution.c -o solution

# Execute the compiled program
./solution
```
