```python
def sumCounts(nums):
    n = len(nums)
    ans = 0

    for i in range(n):
        s = set()
        for j in range(i, n):
            s.add(nums[j])
            ans += len(s) ** 2

    return ans

print(sumCounts([1,2,1]))
```
