```python
import heapq

def dijkstra(graph, start, end):
    pq = [(0, start)]
    dist = {node: float('inf') for node in graph}
    dist[start] = 0

    while pq:
        cost, node = heapq.heappop(pq)

        if node == end:
            return cost

        for neighbor, weight in graph[node]:
            new_cost = cost + weight

            if new_cost < dist[neighbor]:
                dist[neighbor] = new_cost
                heapq.heappush(pq, (new_cost, neighbor))

    return -1

graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('D', 5)],
    'C': [('D', 8), ('E', 10)],
    'D': [('F', 2)],
    'E': [('F', 3)],
    'F': []
}

start = 'A'
end = 'F'

print("Shortest Path Cost =", dijkstra(graph, start, end))
```

```python
import heapq

def schedule_jobs(jobs, machines):

    machine_load = [0] * machines
    heapq.heapify(machine_load)

    for job in sorted(jobs, reverse=True):
        load = heapq.heappop(machine_load)
        heapq.heappush(machine_load, load + job)

    return max(machine_load)

n = int(input("Enter number of jobs: "))
jobs = list(map(int, input("Enter job processing times: ").split()))
machines = int(input("Enter number of machines: "))

print("Minimum Completion Time =", schedule_jobs(jobs, machines))
```
