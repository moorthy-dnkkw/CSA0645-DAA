```python
%%bash
# Write the C code to a file named solution.c using a here document
cat <<EOF > solution.c
#include <stdio.h>

int min(int a,int b){ return a<b?a:b; }

int main() {
    int a1[] = {4,5,3,2};
    int a2[] = {2,10,1,4};
    int t1[] = {7,4,5};
    int t2[] = {9,2,8};
    int e1=10,e2=12,x1=18,x2=7;
    int n=4;

    int f1[n], f2[n];
    f1[0]=e1+a1[0];
    f2[0]=e2+a2[0];

    for(int i=1;i<n;i++){
        f1[i]=min(f1[i-1]+a1[i], f2[i-1]+t2[i-1]+a1[i]);
        f2[i]=min(f2[i-1]+a2[i], f1[i-1]+t1[i-1]+a2[i]);
    }

    printf("%d", min(f1[n-1]+x1, f2[n-1]+x2));
}
EOF

# Compile the C code
gcc -o solution solution.c

# Run the compiled program
./solution
```
