```python
%%bash
cat <<EOF > binary_search.c
#include <stdio.h>

int main() {
    int arr[] = {3,9,14,19,25,31,42,47,53};
    int n = 9;
    int key = 31;
    int low = 0, high = n - 1;

    while (low <= high) {
        int mid = (low + high) / 2;
        printf("Mid index = %d, Value = %d\n", mid, arr[mid]);

        if (arr[mid] == key) {
            printf("Position = %d", mid + 1);
            return 0;
        } else if (arr[mid] < key)
            low = mid + 1;
        else
            high = mid - 1;
    }

    printf("Not Found");
    return 0;
}
EOF
gcc -o binary_search binary_search.c
./binary_search
```
